```python
import gc
import sys
from collections import Counter

# Force a garbage collection
gc.collect()

# Get all tracked objects
objs = gc.get_objects()

# Sort by size (largest first)
objs = sorted(objs, key=sys.getsizeof, reverse=True)

# Print the largest objects
for obj in objs[:20]:
    print(type(obj), sys.getsizeof(obj), repr(obj)[:100])

print(Counter(type(obj) for obj in objs).most_common(20))
```
