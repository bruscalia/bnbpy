# Maximum weighted independent set

This is just a bonus notebook as the heuristic is used in Graph Coloring to produce new columns.

$$
 \begin{align}
     \text{max} \quad & \sum_{i \in N} x_{i} \\
     \text{s.t} \quad & x_{i} + x_{j} \leq 1 & \forall \; i, j \in E \\
 \end{align}
 $$

````python
from bnbprob import gcol
````

````python
# Instance has a list of tuples (edges) in which node indexes start at 0
instance = gcol.load_instance("./../data/gcol/gcol32.txt")
````

````python
heur = gcol.IndepSetHeur()
graph = gcol.IndepGraph(instance["edges"])
sol = heur.solve(graph, save_history=True)
````

````python
gcol.draw_mis_from_nodes(
    graph.nodes,
    active_color='firebrick',
    seed=12,
    use_labels=False,
    plot_margins=False,
)
````
