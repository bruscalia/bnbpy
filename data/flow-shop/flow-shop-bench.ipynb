```python
import json
import re
from pathlib import Path


def parse_instance(data):
    """Parses a single flowshop instance and returns the processing times."""
    lines = data.strip().split('\n')
    jobs, machines = map(int, lines[2].split())
    processing_times = []

    for line in lines[3:-1]:
        times = [int(x) for i, x in enumerate(line.split()) if i % 2 == 1]
        processing_times.append(times)

    return processing_times


def process_flowshop_file(content: str, output_dir: str):

    # Split content into instances using 'instance <name>' pattern
    instances = re.split(r'\sinstance\s(\w+)', content)
    base_dir = Path(output_dir)
    base_dir.mkdir(parents=True, exist_ok=True)

    for i in range(1, len(instances), 2):
        instance_name = instances[i].strip()
        instance_data = instances[i + 1]
        processing_times = parse_instance(instance_data)

        output_path = base_dir / f'{instance_name}.json'
        with open(output_path, 'w', encoding='utf8') as json_file:
            json.dump(processing_times, json_file, indent=4)

        print(
            f"Instance '{instance_name}' processed and saved to {output_path}"
        )


# Example Usage:
# Assuming `flowshop1.txt` is your input file and the current directory
# is the target output directory
process_flowshop_file(inst, '../data/flow-shop')
```

```python
import json
import re
from pathlib import Path


def parse_instances_from_text():
    """
    Parse instance definitions from a Taillard
    benchmark description text file.
    """
    instances = []
    content = tai

    # Regex to match lines with seed and instance names
    instance_regex = r"(\d+)\s+ta(\d{3})"
    matches = re.findall(instance_regex, content)

    # Hardcoded job and machine counts based on sections
    # Adjust if these change
    num_jobs_machines = [
        (20, 5), (20, 10), (20, 20), (50, 5), (50, 10), (50, 20),
        (100, 5), (100, 10), (100, 20), (200, 10), (200, 20), (500, 20)
    ]

    # Distribute matches across jobs and machines
    section_size = 10  # 10 instances per section in the input text
    for idx, (num_jobs, num_machines) in enumerate(num_jobs_machines):
        start_idx = idx * section_size
        end_idx = start_idx + section_size
        for match in matches[start_idx:end_idx]:
            seed, name_suffix = match
            instances.append({
                "seed": int(seed),
                "num_jobs": num_jobs,
                "num_machines": num_machines,
                "name": f"ta{name_suffix}"
            })

    return instances


def unif(seed, low, high):
    """Replicate Pascal/C uniform random number generator."""
    m = 2147483647
    a = 16807
    b = 127773
    c = 2836
    k = seed // b
    seed = a * (seed % b) - k * c
    if seed < 0:
        seed += m
    value_0_1 = seed / m
    return seed, low + int(value_0_1 * (high - low + 1))


def generate_processing_times_transposed(
    seed, num_jobs, num_machines, low=1, high=99
):
    """Generate processing times using Pascal/C random logic."""
    times = [[] for _ in range(num_jobs)]  # Transposed: list for each job
    for i in range(num_machines):
        for j in range(num_jobs):
            seed, value = unif(seed, low, high)
            times[j].append(value)  # Append to the respective job's list
    return seed, times


def generate_and_save_all_instances(instance_list, output_dir):
    """
    Generate processing times for all instances and
    save them to JSON files.
    """
    output_dir.mkdir(exist_ok=True)
    for instance in instance_list:
        # Generate processing times in transposed format
        _, processing_times = generate_processing_times_transposed(
            instance['seed'], instance['num_jobs'], instance['num_machines']
        )
        # Save to JSON
        output = {
            'instance': instance['name'],
            'num_jobs': instance['num_jobs'],
            'num_machines': instance['num_machines'],
            'processing_times': processing_times,
        }
        file_path = output_dir / f"{instance['name']}.json"
        with open(file_path, mode='w', encoding='utf8') as json_file:
            json.dump(processing_times, json_file, indent=4)
        print(f"Saved {instance['name']} to {file_path}")


output_directory = Path('../data/flow-shop')

# Parse instances and generate JSONs
try:
    # Parse instances from the uploaded text file
    instance_definitions = parse_instances_from_text()

    # Generate JSON files for all instances
    generate_and_save_all_instances(instance_definitions, output_directory)

except Exception as e:
    str(e)
```
